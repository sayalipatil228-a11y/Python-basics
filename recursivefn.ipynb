# Python Recursive Function


```python
def factorial(num):
    if (num == 0 or num == 1):
        return 1
    else:
        return num * factorial(num -1)

print(factorial(3))
    
# num = 7;
# print("Number:",num)
# print("Factorial:", factorial)
```
