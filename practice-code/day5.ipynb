## üîπ 1Ô∏è‚É£ What is a String?

A string is text written inside quotes.

### üîπ 2Ô∏è‚É£ String Indexing (Very Important ‚≠ê)

Each character has a position (index).

```python
word = "Python"
print(word[0])
print(word[5])
```

#### üî∏ Negative Indexing

```python
word = "Python"
print(word[-2])
print(word[-4])
```

### üîπ 3Ô∏è‚É£ String Slicing

```python
text = "Programming"
print(text[3:6])
print(text[0:11:2])
print(text[:6])
print(text[3:])
print(text[:])
```

## üîπ 4Ô∏è‚É£ String Length

```python
name = "Sayali"
print(len(name))
```

## üîπ 5Ô∏è‚É£ Common String Methods ‚≠ê‚≠ê


```python
msg = "Hello, python!"
print(msg)
print(msg.upper())
print(msg.lower())
print(msg.capitalize())
print(msg.title())
print(msg.strip())
print(msg.replace("python", "world"))
print(msg.find("hello"))
```

## üîπ 6Ô∏è‚É£ String Concatenation (Join Strings)

```python
first = "Sayali"
last = "Patil"
full = first + " " + last
print(full)
```

## üîπ 7Ô∏è‚É£ f-String (Best Way ‚≠ê)

```python
name = "Sayali"
age = 25
print(f"My Name is {name} & I am {age} years old.")
```
