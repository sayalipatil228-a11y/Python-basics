```python
class BankAccount:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        return self.balance

    def withdraw(self, amount):
        if amount > self.balance:
            raise ValueError("Insufficient Balance")
        else:
            self.balance -= amount
            return self.balance
    
    def get_balance(self):
        return self.balance

#-----------------Main Program---------------------------

with open("account.txt", "r") as file:
    data = file.read().strip()
    name, balance = data.split(",")

acc = BankAccount(name, int(balance))

print("Name:", acc.name)
print("Balance:", acc.get_balance())

choice = input("Do you want to (withdraw / deposit) :").lower()

if choice == "withdraw":
    try:
        amount = int(input("Engter Amount to Withdraw:"))
        new_balance = acc.withdraw(amount)
        print("Withdrawal successful")
    except ValueError as e:
            print("Error:", e)


elif choice == "deposit":
    amount = int(input("Engter Amount to Deposit:"))
    new_balance = acc.deposit(amount)
    print("Deposit successful")

else:
    print("Invalid Choice")
    new_balance = acc.get_balance()

# UPDATED FILE
with open("account.txt", "w") as file:
    file.write(f"{acc.name},{new_balance}")

print("Updated Balance:", new_balance)


   
       
```
