```python
student = {
    "name": "Sayali",
    "age": 20,
    "course": "Python"
}

for key, value in student.items():
    print(key, ":", value)
```

### 4Ô∏è‚É£ Write a Small Class

A class groups data + functions.

    üîë __init__ runs automatically
    üîë self refers to current object

Example:

```python

class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def show(self):
        print(self.name, self.age)

s1 = Student("Sayali", 20)
s1.show()
```
