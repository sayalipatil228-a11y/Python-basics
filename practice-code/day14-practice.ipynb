    result() â†’ Pass if marks â‰¥ 40 else Fail

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
    
    def display(self):
        print("Student Name:", self.name)
        print("Marks:", self.marks)
    
    def result(self):
        if self.marks >= 90:
            print("grade: A")
        elif self.marks >= 70:
            print("grade: B")
        elif self.marks >= 50:
            print("Result: Pass")
        else:
            print("Result: Fail")

s1 = Student("Saya", 98)

s1.display()
s1.result()

```

#### ðŸ§© Project 2: Bank Account (Improved)
Requirements

    Class BankAccount

Methods:

    deposit

    withdraw

    show_balance

Rule:

Raise error if withdraw > balance

```python
class BankAccount:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance 

    def show_balance(self):
        print("Name:", self.name)
        print("Total balance:", self.balance)
        

    def deposit(self, amount):
        self.balance += amount
        print("Deposit:", amount)
        print("New Balance:", self.balance)

    def withdraw(self, amount):
        if amount > self.balance:
            print("Error: Insufficient balance")
        else:
            self.balance -= amount
            print("Withdraw:", amount)
            print("Remaining Balance:", self.balance)

acc1 = BankAccount("Saya", 20000)
acc1.show_balance()
acc1.deposit(20000)
acc1.withdraw(2000)
```

#### ðŸ§© Project 3: Shape Area Calculator (Polymorphism)
Requirements

        Parent class Shape

Child classes:

        Rectangle

        Circle

Same method name area()

```python
class Shape:
    def __init__(self):
        pass

class Rectangle(Shape):
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def area (self):
        return self.length * self.breadth
    
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area (self):
        return 3.14 * self.radius * self.radius        

shapes = [Rectangle(30, 6), Circle(5)]

# Polymorphism in action
for shape in shapes:
    print("Area :", shape.area())
```
