## üîÅ **What is a Loop?**
A loop runs the same code multiple times.
Instead of writing print() 5 times ‚Üí use a loop üëç
### üîπ 1Ô∏è‚É£ for Loop (Most Used)

```python
for i in range(5):
    print(i)
```

```python
for i in range(1, 6):
    print(i)
```

### üîπ 2Ô∏è‚É£ while Loop

Runs as long as condition is True

```python
i = 1
while i <= 5:
    print(i)
    i = i + 1
```

### üîπ 3Ô∏è‚É£ break Statement

Stops the loop

```python
for i in range(1, 10):
    if i == 5:
        break
    print(i)
```

### üîπ 4Ô∏è‚É£ continue Statement
Skips current iteration

```python
for i in range(1, 11):
    if i == 5:
        continue
    print(i)
```

### üîπ 5Ô∏è‚É£ pass Statement

Does nothing (used as placeholder)

```python
for i in range(5):
    pass
```

##### Task 2Ô∏è‚É£ ---Print even numbers from 1 to 20

```python
for i in range (1, 21):
    if i % 2 == 0:
        print(i)
#  even numbers only
```

```python
# or 
for i in range(1, 21):
    if i % 2 != 0:
        continue
    print(i)
# even numbers 
```

```python
# odd numbers
for i in range(1, 10):
    if i % 2 != 0:
        print(i)
```
