### üîπ 1Ô∏è‚É£ What is a Dictionary?

A dictionary stores data in key : value pairs.

```python
student = {
    "name" : "John Doe",
    "age" : 21,
    "courses" : ["Math", "CompSci"]
}
print(student)
```

üî∏Keys are unique

üî∏Values can be anything

üî∏Unordered

üî∏Mutable (can change)

### üîπ 2Ô∏è‚É£ Accessing Values

```python
print(student["name"])
print(student["age"])
```

#### üîπOr using .get()

```python
print(student.get("course"))
```

### üîπ 3Ô∏è‚É£ Changing & Adding Data

```python
student["age"] = 22 # Update existing key
student["city"] = "New York" # Add new key-value pair
print(student)
```

### üîπ 4Ô∏è‚É£ Removing Data

```python
student = {
    "name" : "John Doe",
    "age" : 21,
    "courses" : ["Math", "CompSci"]
}
print(student)
student.pop("age")
del student["courses"]
print(student)
```

### üîπ 5Ô∏è‚É£ Dictionary Methods

```python
print(student.keys())
print(student.values())
print(student.items())
```

### üîπ 6Ô∏è‚É£ Loop Through Dictionary

```python
for key in student:
    print(key, student[key])
```

```python
# or 
for k, v in student.items():
    print(k, ":", v)
```

##### ‚úÖ Task 2 ---Change the marks and print updated dictionary.

```python
student = {
    "name": "John Doe",
    "age": 21,
    "marks": "[85, 90, 78]"
}
student["marks"] = [59, 85, 88]  # update the marks 
for k, v in student.items():
    print(k, ":", v)
```

##### ‚úÖ Task 3 ---Add a new key grade and delete age.

```python
student = {
    "name": "John Doe",
    "age": 21,
    "marks": "[85, 90, 78]"
}
student["grade"] = "A"
del student["age" ]
for k, v in student.items():
    print(k, ":", v)
```
