## üîπ 1Ô∏è‚É£ What is Inheritance?

Inheritance means:

    Child class uses properties & methods of Parent class

Real life:

Parent ‚Üí Person

Child ‚Üí Student, Employee

### üîπ 2Ô∏è‚É£ Parent Class Example

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def show(self):
        print("Name:",self.name)
        print("Age:", self.age)

p1 = Person("abc", 59)
p1.show()        
```

### üîπ 3Ô∏è‚É£ Child Class (Inheritance)

```python
class Student(Person):
    def __init__(self, name, age, marks):
        super().__init__(name, age)
        self.marks = marks
    
    def show_Student(self):
        self.show()
        print("Marks:", self.marks)

s1 = Student("Saya", 21, 90)
s1.show_Student()
```

#### üîπ 4Ô∏è‚É£ Why super() is IMPORTANT

    super().__init__(name, age)

It means:

    ‚ÄúCall parent‚Äôs constructor‚Äù

Without super() ‚Üí parent variables won‚Äôt exist.

#### üîπ 5Ô∏è‚É£ Another Example (Employee)

```python
class Employee(Person):
    def __init__(self, name, age, salary):
        super().__init__(name, age)
        self.salary = salary
    
    def show_Employee(self):
        self.show()
        print("Salary:", self.salary)

e1 = Employee("xyz", 32, 30000)
e1.show_Employee()
```

#### üîπ 6Ô∏è‚É£ Polymorphism (Same method, different behavior)

```python
class Animal:
    def sound(self):
        print("Animal Makes Sound")
    
class Dog(Animal):
    def sound(self):
        print("Dog Barks")

class Cat(Animal):
    def sound(self):
        print("Cat Meows")
    
animals = [Dog(), Cat(), Animal()]
for a in animals:
    a.sound()
        
    
```

Create class Car that inherits Vehicle

```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def show_brand(self):  
        print("Vehicle Brand:", self.brand)

class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model

    def show_car(self):
        self.show_brand()
        print("Car Model:", self.model)

c1 = Car("Land Rover","Range Rover")
c1.show_car()
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def show_person(self):
        print("Person Name is:", self.name)
        print("Age is:", self.age)

class Employee(Person):
    def __init__(self, name, age, salary):
        super().__init__(name, age)
        self.salary = salary

    def show_employee(self):
        self.show_person()
        print("Salary :", self.salary)

p1 = Employee("Conrad", 26, 60000)
p1.show_employee()

       
        
```

Override area() (polymorphism).

```python
class Shape:
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def area(self):
        return self.length * self.breadth

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius


shapes = [Rectangle(10, 7), Circle(5)]

for Shape in shapes:
    print("Area:", Shape.area())        
```
