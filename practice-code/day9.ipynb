#### âœ… Task 4 ---Create a function to check if a number is even or odd.

```python
def eod(num):
    if num % 2 == 0:
        return True
    else:
        return False
    
print(eod(4))  # True
print(eod(7))  # False
```

#### âœ… Task 5 â­ ---Create a function that takes 3 numbers and returns the largest.

```python
def larger(a, b, c):
    if a > b and a > c:
        return a
    elif b > a and b > c:
        return b
    else:
        return c
print("Largest:", larger(10, 20, 15))
```
