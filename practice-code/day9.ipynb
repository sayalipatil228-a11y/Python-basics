#### üîπ Task 8 ‚≠ê ---Create a function calculator(a, b, op)
If op is:

"+" ‚Üí add

"-" ‚Üí subtract

"*" ‚Üí multiply

"/" ‚Üí divide

calculator(10, 5, "+") ‚Üí 15
calculator(10, 5, "*") ‚Üí 50




```python
def  calculator(a, b, op):
    if op == '+':
        return a + b
    elif op == '-':
        return a - b
    elif op == '*':
        return a * b
    elif op == '/':
        return a / b
    else:
        return "Invalid operator"
operation = input("Enter any operation (+, -, *, /): ")
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
result = calculator(a, b, operation)
print("Result:", result)
```
