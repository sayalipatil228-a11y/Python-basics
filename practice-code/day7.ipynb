### üîπ 1Ô∏è‚É£ Tuples (Immutable Lists) ‚Äì 45 min
**What is a Tuple?**

A tuple is like a list, but cannot be changed.

‚úî Ordered

‚úî Immutable

‚úî Allows duplicates

```python
t = (12, 34, 56)
print(t)
```

### **Access Tuple Elements**

```python
t = (12, 34, 56)
print(t[0])
print(t[1-2])
print(t[-1])
```

**Why Tuples?**

Faster than lists

Used for fixed data

Protects data from change

### **Tuple Methods**

```python
t = (1, 2, 3, 4)
print(t.count(2))
print(t.index(4))
```

## üîπ 2Ô∏è‚É£ Sets (Unique & Unordered) 
### **What is a Set?**

A set stores unique values only.

‚ùå No duplicates

‚ùå No indexing

‚úî Fast operations

```python
s = {1, 2, 3, 4, 5}
print(s) 
```

### Remove Duplicates Using Set

```python
num = [1,2,2,3,3,4,5]
unique_num = set(num)
print(unique_num)
```

#### **Set Methods**

```python
a = {1, 2, 3}
b = {3, 4, 5}
print(a.union(b)) 
print(a.intersection(b))
print(a.difference(b))
```
