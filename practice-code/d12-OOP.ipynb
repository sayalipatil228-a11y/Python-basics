### üîπ 5Ô∏è‚É£ Constructor (__init__)

This runs automatically when object is created.

```python
class student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

s1 = student("Alice", 85)
s2 = student("Bob", 90)
print(s1.name, s1.marks)
print(s2.name, s2.marks)
```

### üîπ 6Ô∏è‚É£ Methods (Functions inside class)

```python
class stud:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def show(self):
       print(self.name, self.marks)

s1 = stud("Charlie", 95)
s1.show()
```
