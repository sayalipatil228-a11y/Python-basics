#### 1Ô∏è‚É£ Student Report System

    Create a class Student:

name

marks

Methods:

display()

is_pass() ‚Üí returns True if marks ‚â• 40

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print("Name:", self.name)
        print("Marks:", self.marks)

    def is_passed(self):
        if self.marks >= 40:
            return True
        else:
            return False
    
s1 = Student("maya", 38)
s1.display()

if s1.is_passed():
    print("Passed")
else:
    print("Failed")
    
```

#### 2Ô∏è‚É£ Mobile Phone

    Create a class Phone:

brand

price

Methods:

    show()

    discount(percent)

```python
class Phone:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

    def show(self):
        print("Brand:", self.brand)
        print("Price:", self.price)

    def discount(self, percent):
        discount_amount = self.price * (percent / 100)
        self.price -= discount_amount

        print("New Price after discount:", self.price)

phone1 = Phone("Apple", 1000)
phone1.show()
phone1.discount(10)
phone1.show()


        
```

#### 3Ô∏è‚É£ Bank Account

    Class BankAccount

name   balance

Methods:

    deposit(amount)

    withdraw(amount)

    show_balance()

Rules:

If withdraw > balance ‚Üí show error.

```python
class BankAccount:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance  
    
    def deposit(self, amount):
        self.balance += amount
        print("Deposited:", self.balance)

    def withdraw(self, amount):
        if amount > self.balance:
            print("Error: Insufficient funds")
        else:
            self.balance -= amount
            print("Withdrawn:", amount) 

    def show_balance(self):
        if self.balance < 0:
            print("Error: Insufficient funds")
        else:
            print(self.name, "Your balance is:", self.balance)

acc = BankAccount("saya", 10000000)
acc.show_balance()
acc.deposit(5000000)
acc.show_balance()

        
```

#### 4Ô∏è‚É£ ATM System

    Class ATM

pin

balance

Methods:

    check_pin()

    withdraw()

    check_balance()

```python
class ATM:
    def __init__(self, pin, balance):
        self.pin = pin
        self.balance = balance

    def check_pin(self, entered_pin):
        return self.pin == entered_pin
    
    def check_balance(self):
        print("Current balance:", self.balance)
    
    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance")
        else:
            self.balance -= amount
            print("Withdrawn :", amount)
            print("Remaining balance:", self.balance)

atm = ATM(1234, 200000)

user_pin = int(input("Enter your pin:"))

if atm.check_pin(user_pin):
    print("Valid pin")
    atm.check_balance()
    amount = int(input("Enter amount to withdraw:"))
    atm.withdraw(amount)
else:
    print("Invalid Pin")
    
        
```
