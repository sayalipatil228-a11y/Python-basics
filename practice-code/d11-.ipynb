    Step 1 â€“ Normal program (can crash)

```python
x = int(input("Enter number: "))
y = int(input("Enter another number: "))

print(x/ y)
```

#### Step 2 â€“ Using try & except
    Now the program does not crash ðŸŽ‰

```python
try:
    x = int(input("Enter number: "))
    y = int(input("Enter another number: "))

    print(x/ y)

except:
    print("something went wrong")
```

### Step 3 â€“ Catch specific errors
    Now Python tells what went wrong.

```python
try:
    x = int(input("Enter number: "))
    y = int(input("Enter another number: "))

    print(x/ y)

except ZeroDivisionError:
    print("You can't divide by zero!")

except ValueError:
    print("Invalid input. Please enter a valid number.")
```

#### Step 4 â€“ finally
    finally runs always, error or not.

```python
try:

    print(10/ 0)

except :
    print("Error occurred.")

finally:
    print("Program finished.")
```

### ðŸ”¹ PART 2 â€” Raising Custom Errors
    This lets you decide when to stop a program.

```python
age = int(input("Enter age: "))

if age < 18:
    raise ValueError("You are under 18")

print("Access granted")
```

#### ðŸ”¹ PART 3 â€” enumerate()
    This is used everywhere in real projects.

```python
names = ["Alice", "Bob", "Charlie"]

for index, name in enumerate(names, start=1):
    print(index, name)
```
