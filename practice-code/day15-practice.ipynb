üéØ REQUIREMENTS (READ CAREFULLY)
    Class: ATM

Attributes:

    pin (int)

    balance (int)

    attempts (int, start at 0)

Methods:

    verify_pin(entered_pin)

Returns True or False

Increments attempts on failure

    withdraw(amount)

If amount > balance ‚Üí raise ValueError

Else subtract and return new balance

```python
class ATM:
    def __init__(self, pin, balance):
        self.pin = pin 
        self.balance = balance
        self.attempts = 0
    
    def verify_pin(self, entered_pin):

        if entered_pin == self.pin:
            return True
        else:
            self.attempts += 1
            return False

    def withdraw(self, amount):
        if amount > self.balance:
            raise ValueError("Insufficient balance")
        self.balance -= amount
        return self.balance
        

atm = ATM(1234, 200000)

while atm.attempts < 3:
    try:
        entered_pin =  int(input("Enter Pin:"))

        if atm.verify_pin(entered_pin):
            amount = int(input("Enter amount to withdraw:"))
            new_balance = atm.withdraw(amount)
            print("withdrawal successful")
            print("Remaining Balance:", new_balance)
            break
        else:
            print("Wrong Pin")

    except ValueError as e:
        print("Error :", e)

if atm.attempts == 3:
    print("Card Blocked due to 3 wrong attempts ")



```

üîÅ OTHER EXAMPLE (Same Discipline, Different Context)
Example: Door Lock System
Idea:

Lock has a password

3 wrong tries ‚Üí locked

Same principles as ATM

```python
class DoorLock:
    def __init__(self, password):
        self.password = password
        self.attempts = 0

    def unlock(self, entered_password):
        if entered_password == self.password:
            return True
        else:
            self.attempts += 1
            return False
        
# ----------Main Program---------------------
lock = DoorLock("123")

while lock.attempts < 3:
    entered = input("enter password:")

    if lock.unlock(entered):
        print("Door Unlocked!")
        break
    else:
        print("Wrong Password")

if lock.attempts == 3:
        print("Door permanently locked")        
```
