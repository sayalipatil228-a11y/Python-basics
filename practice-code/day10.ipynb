#### üîπ 2Ô∏è‚É£ Writing to a File

This creates a file called data.txt.

```python
f = open("example.txt", "w")
f.write("namaste doston \n")
f.write("kya hal chal \n")
f.close()  
```

#### üîπ 3Ô∏è‚É£ Reading a File

```python
f = open("example.txt", "r")
content = f.read()
print(content)
f.close()
```

#### üîπ 4Ô∏è‚É£ Append (Add without deleting old data)

```python
f = open("example.txt", "a")
f.write("\n New line added.")
f.close()
```

#### üîπ 5Ô∏è‚É£ Best Practice (Using with)

**Auto closes file ‚úîÔ∏è**

```python
with open("example.txt", "w") as f:
    f.write("Namaskaram \n ")

with open("example.txt", "r") as f:
    print(f.read())
```

## üî• Mini Project ‚Äì Student Record System

#### Create a program that:

Asks user for name, marks

Saves it to students.txt

Reads & displays all student records

```python
with open("students.txt", "a") as f:
    name = input("Enter student name: ")
    marks = input("Enter student marks: ")
    f.write(name + "-" + marks + "\n")

print("All Students Data: ")
with open("students.txt", "r") as f:
    print(f.read())
```
