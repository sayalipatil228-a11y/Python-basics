### ðŸ”¹ **1ï¸âƒ£ What is a List?**

A list stores multiple values in one variable.

```python
numbers = [1, 2, 3, 4, 5]
print(numbers)
name = ["Sayali","Isha", "Pranali", "Mitali", "Aishwarya"]
print(name)
```

### ðŸ”¹ 2ï¸âƒ£ Access List Items (Indexing)

```python
name = ["Sayali","Isha", "Pranali", "Mitali", "Aishwarya"]
print(name[0])
print(name[-1])
```

### ðŸ”¹ 3ï¸âƒ£ Change List Items

```python
name = ["Sayali","Isha", "Pranali", "Mitali", "Aishwarya"]
name[0] = "Rahul"
print(name[0])
print(name[-1])
```

### ðŸ”¹ 4ï¸âƒ£ List Length

```python
name = ["Sayali","Isha", "Pranali", "Mitali", "Aishwarya"]
print(len(name))
```

### ðŸ”¹ 5ï¸âƒ£ Add Items to List â­
**append() â†’ add at end**

```python
name = ["Isha", "Pranali", "Mitali"]
name.append("Ayesha")
print(name)
```

**insert() â†’ add at position**

```python
name = ["Isha", "Pranali", "Monika"]
name.insert(1,"Harish")
print(name)
```

### ðŸ”¹ 6ï¸âƒ£ Remove Items from List â­

```python
name = ["Isha", "Pranali", "Mitali"]
name.remove("Isha")
print(name)
```

```python
name = ["Isha", "Pranali", "Mitali", "Shubham"]
name.pop()
name.pop(0)
print(name)
```

### ðŸ”¹ 7ï¸âƒ£ Loop Through List

```python
names = ["Isha", "Pranali", "Mitali", "Shubham"]
for name in names:
    print(name)
```

```python
numbers = [1, 2, 3, 4, 5]
for number in numbers:
 print(number)
```

### ðŸ”¹ 8ï¸âƒ£ Check Item Exists

```python
names = ["Isha", "Pranali", "Mitali", "Shubham"]
if "Isha" in names:
    print("Hello " + "Isha")
```

### ðŸ”¹ 9ï¸âƒ£ Useful List Methods

```python
nums = [1, 2, 3, 4, 5]
nums.sort()
print(nums)
nums.reverse()
print(nums)
```
